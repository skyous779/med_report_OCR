# merge_合并每一个case的多个xlsx文件为一个csv文件

## main procession

```python
import os
import glob
import pandas as pd
# save_dir = '/home/data1/skyous/gzyy_jc_sum_0330'
save_dir = '/home/skyous/git/ocr_table/debug_merge_table2'
os.makedirs(save_dir, exist_ok=True)

table_dir = "./gzyy_check_list"
person_dir = glob.glob(os.path.join(table_dir, "*"))

# person_dir
```

```python
def delete_empty_rows(df):
    # todo 删除空行
    empty_columns = df.columns[df.isna().all()]
    # 对空列右边的所有列向左平移一列
    if len(empty_columns) > 0:
        for column in reversed(empty_columns):
            if column not in ['姓名', '患者编号', '性别', '采集日期', '出生日期', '年龄']:
                # 找到空列的索引
                empty_column_index = df.columns.get_loc(column)
                # 向左平移一列
                df.iloc[:, empty_column_index:-1] = df.iloc[:, empty_column_index + 1:]
                # 删除最后一列的表头
                df = df.drop(df.columns[-1], axis=1)
    return df

def prepossessing(df):

    columns_list = df.columns.tolist()
    if '年龄' not in columns_list:
        df['年龄'] = None
    df[['姓名', '患者编号', '性别', '采集日期', '出生日期', '年龄']] = df[['姓名', '患者编号', '性别', '采集日期', '出生日期', '年龄']].ffill()
    df = delete_empty_rows(df)

    return df
```

```python

for person_path in person_dir:
    person_name = person_path.split('/')[-1]
    save_path = os.path.join(save_dir, person_name + '.csv')
    print(person_path)

    # save_folder = os.path.join(save_dir, person_name)
    # os.makedirs(save_folder, exist_ok=True)

    person_tables = glob.glob(os.path.join(person_path, "*.xlsx"))

    # 进行排序
    person_tables = sorted(person_tables, key=lambda x: int(x.split("/")[-1].split("_")[2]))


    df_list = []
    for table_path in person_tables:
        # print(table_path)
        df = pd.read_excel(table_path)
        df = prepossessing(df)
        df_list.append(df)
        # image_name = image_path.split('/')[-1]
        # save_path = os.path.join(save_folder, image_name)
        # os.system('cp {} {}'.format(image_path, save_path))

    # 合并多个 DataFrame，表头取并集，缺乏的内容填充为 NaN
    if df_list != []:
        merged_df = pd.concat(df_list, axis=0, join='outer', ignore_index=True)
        merged_df.to_csv(save_path, index=False)


```
